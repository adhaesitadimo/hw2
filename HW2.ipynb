## 5. Подсчет метрик для графа

Рассчитаем для графа следующие метрики: диаметр (diameter), эксцентриситет (eccentricity), периферию (periphery). В принципе, как считать эти метрики, описано в комментариях и в коде.

```python
def count_eccentricity(graph):
    # count eccentricity for every node
    # eccentricity is the maximum shortest path for each node to any node
    eccentricities = []
    for node in range(graph.number_of_nodes()):
        length = nx.single_source_shortest_path_length(graph, node)
        eccentricities.append(max(length.values()))
    return eccentricities

def count_diameter(graph):
    # diameter is the maximum eccentricity of a graph
    return(max(count_eccentricity(graph)))

def count_periphery(graph):
    # periphery is the number of nodes with eccentricity equal to diameter
    ecc = count_eccentricity(graph)
    d = count_diameter(graph)
    return [i for i in range(graph.number_of_nodes()) if ecc[i] == d]
 
```

```python
diameter1 = count_diameter(terror2002)
diameter2 = count_diameter(terror2003)
```

```python
ecc1 = np.mean(count_eccentricity(terror2002))
ecc2 = np.mean(count_eccentricity(terror2003))
```

Наши графы невзвешены, поэтому диаметр и эксцентриситет считаются в числе ребер от точки до точки. Поскольку графы 2002 и 2003 года различных размеров, поможет нормировка по общему числу ребер.

```python
diameter1 = (diameter1 / terror2002.number_of_edges()) * 100
diameter2 = (diameter2 / terror2003.number_of_edges()) * 100

print(diameter1, diameter2, abs(diameter1 - diameter2))
```

```python
ecc1 = (ecc1 / terror2002.number_of_edges()) * 100
ecc2 = (ecc2 / terror2003.number_of_edges()) * 100

print(ecc1, ecc2, abs(ecc1 - ecc2))
```

Назовем выведенные величины "относительным диаметром" и "относительным средним эксцентриситетом" (в процентах). Мы видим, что у первого графа относительный эксцентриситет и относительный эксцентриситет ниже на 4 и 3.5 процента соответственно. Это говорит о том, что в среднем пути в первом графе несколько короче, чем во втором, хотя разница 3-4 % говорит, что в этом плане графы достаточно похожи.

```python
per1 = len(count_periphery(terror2002))
per2 = len(count_periphery(terror2003))
```

Периферия оценивает количество вершин с максимальным эксцентриситетом, поэтому тут уже логичнее нормировать по числу вершин

```python
per1 = (per1 / terror2002.number_of_nodes()) * 100
per2 = (per2 / terror2003.number_of_nodes()) * 100

print(per1, per2, abs(per1 - per2))
```

Здесь мы видим, что в обоих графах довольно значительный процент вершин находится "на периферии", при этом здесь графы уже сильно "похожи" - процент почти одинаковый.

## 6. Подсчет Wiener index

Подсчитаем индекс Винера - сумму длин всех самых длинных путей в графе.

```python
def wiener_index(graph):
    # Wiener index is the sum of all shortest path lengths between all nodes in graph
    return sum([sum(item[1].values()) for item in nx.shortest_path_length(graph)])
```

```python
wiener1 = wiener_index(terror2002)
wiener2 = wiener_index(terror2003)
```

Рассмотрим случай полного графа. В нем все вершины связаны со всеми, поэтому от любой до любой вершины мы сможем добраться ровно через одно ребро. Это минимально возможный кратчайший путь, тогда и сумма по индексу Винера будет минимальна. Это будет "идеальный" граф. Всего ребер в полном графе n(n-1), где n - число вершин.

Нормировать будем по этому "идеальному" случаю - будем рассматривать отношения минимального идекса Винера к получившемуся.

```python
wiener1 = (terror2002.number_of_nodes() * (terror2002.number_of_nodes() - 1)) / wiener1
wiener2 = (terror2003.number_of_nodes() * (terror2003.number_of_nodes() - 1)) / wiener2

print(wiener1, wiener2, abs(wiener1 - wiener2))
```

Оба графа довольно далеки от идеального, при этом первый ближе. Разница между ними по этой метрике - 12 процентов.

## 8. Подсчет ядра графа

Здесь нужно найти множество внутренней устойчивости графов. По определению это максимальное множество таких вершин, ни одна из которых попарно не смежна.

Здесь можно очень хитро воспользоваться топологией графа и заметить, что ядро можно составить путем выбора по одной вершине из клики, не инцидентных с мостами.

Докажем это утверждение от противного. Пусть мы можем увеличить ядро. Это можно сделать за счет какой-то еще вершины. Но по построению имеем, что все вершины уже состоят в какой-то из клик. Получаем, что новая вершина будет смежная с какой-то уже имеющейся вершиной. Получили противоречие, ч.т.д.

Для того, чтобы получить ядро, выберем по одной вершине из каждой клики таким образом, чтобы она не была инцидентна с мостом (иначе может так получится, что она будет смежна с вершиной другой клики)

```python
kernel2002 = []
for clique in cliques2002:
    i = 0
    while True:
        if clique[i] not in bridge_points_2002:
            kernel2002.append(clique[i])
            break
        else:
            i += 1
            continue
print(kernel2002)

kernel2003 = []
for clique in cliques2003:
    i = 0
    while True:
        if clique[i] not in bridge_points_2003:
            kernel2003.append(clique[i])
            break
        else:
            i += 1
            continue
print(kernel2003)
```
